```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python

def calculate_info(school_data_complete):
    to_enter=[]

    to_enter.append(len(school_data_complete['school_name'].unique()))

    to_enter.append(len(school_data_complete['student_name']))

    to_enter.append(school_data_complete['budget'].sum())

    to_enter.append(school_data_complete['math_score'].mean())

    to_enter.append(school_data_complete['reading_score'].mean())

    students_passed_maths=school_data_complete[school_data_complete['math_score'] >= 70]['student_name'].count()
    to_enter.append((students_passed_maths/len(school_data_complete['student_name']))*100)

    students_passed_reading=school_data_complete[school_data_complete['reading_score'] >= 70]['student_name'].count()
    to_enter.append((students_passed_reading/len(school_data_complete['student_name']))*100)

    total_passed_students=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]['student_name'].count()
    to_enter.append((total_passed_students/len(school_data_complete['student_name']))*100)
    return to_enter
```

```python
df = pd.DataFrame(columns=['total_schools','total_students','total_budget','maths_avg','reading_avg','math_pass','reading_pass','overall_pass'])
df.loc[0]=calculate_info(school_data_complete)
```

```python
df = pd.DataFrame(columns=['school_name','school_type','total_students','total_budget','maths_avg','reading_avg','math_pass','reading_pass','overall_pass'])

for i in range(0,len(schools)):
    temp_df=school_data_complete.loc[school_data_complete['school_name'] == schools[i]]
    df.loc[i]=calculate_info_2(temp_df)
    
df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python

def calculate_info_3(school_data_complete,label):
    to_enter=[]
    to_enter.append(label)

    to_enter.append(school_data_complete['math_score'].mean())

    to_enter.append(school_data_complete['reading_score'].mean())

    students_passed_maths=school_data_complete[school_data_complete['math_score'] >= 70]['student_name'].count()
    to_enter.append((students_passed_maths/len(school_data_complete['student_name']))*100)

    students_passed_reading=school_data_complete[school_data_complete['reading_score'] >= 70]['student_name'].count()
    to_enter.append((students_passed_reading/len(school_data_complete['student_name']))*100)

    total_passed_students=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]['student_name'].count()
    to_enter.append((total_passed_students/len(school_data_complete['student_name']))*100)
    return to_enter
```

```python
df = pd.DataFrame(columns=['Spending Ranges (Per Student)','average math score','Average Reading Score','% passing math','% passing reading','overall passing'])

temp_df=school_data_complete.loc[school_data_complete['budget']/school_data_complete['size'] < 586]
df.loc[0]=calculate_info_3(temp_df,"<$585")

temp_df=school_data_complete.loc[(school_data_complete['budget']/school_data_complete['size']> 584) & (school_data_complete['budget']/school_data_complete['size']<631)]
df.loc[1]=calculate_info_3(temp_df,"$585-630")

temp_df=school_data_complete.loc[(school_data_complete['budget']/school_data_complete['size']> 629) & (school_data_complete['budget']/school_data_complete['size']<646)]
df.loc[2]=calculate_info_3(temp_df,"$630-645")


temp_df=school_data_complete.loc[school_data_complete['size']>644]
df.loc[3]=calculate_info_3(temp_df,"$645-680")
    
df
```
